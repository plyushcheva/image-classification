### Dopředná neuronová síť

FNN jsou vhodné pro úlohy, kde není třeba zachytit prostorové nebo časové závislosti, tedy například u klasifikace na základě statických, předzpracovaných vstupů. Pro dataset s obrázky bez předzpracování není to ideální model.

Uděláme model dopředné neuronové sítě:

````python
class FFNN(nn.Module):
    def __init__(self):
        super(FFNN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.net(x)
````

Funkce pro trénování modelu:

````python
def train_model(model, train_loader, val_loader, epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(epochs):
        model.train()
        running_loss = 0
        total_samples = 0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * xb.size(0)
            total_samples += xb.size(0)

        avg_loss = running_loss / total_samples
        print(f"Epoch {epoch+1}, Train Loss: {avg_loss:.4f}")

        model.eval()
        correct, total = 0, 0
        val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                outputs = model(xb)
                loss = criterion(outputs, yb)
                val_loss += loss.item() * xb.size(0)

                _, predicted = torch.max(outputs.data, 1)
                total += yb.size(0)
                correct += (predicted == yb).sum().item()
        val_accuracy = correct / total
        val_loss = val_loss / total
        print(f"Validation Loss: {val_loss:.4f}, Accuracy: {val_accuracy:.4f}")

model = FFNN()
train_model(model, train_loader, val_loader, epochs=15)
````

Teď zkusme najít vhodnou hloubku a šířku, otestujeme různé možnosti.

````python
def test_architectures(train_loader, val_loader, architectures, epochs=15):
    results = {}
    
    for name, layers in architectures.items():
        print(f"\nTesting architecture: {name}")
        print("-" * 50)
        
        class CustomFFNN(nn.Module):
            def __init__(self):
                super(CustomFFNN, self).__init__()
                self.net = nn.Sequential(*layers)
            
            def forward(self, x):
                return self.net(x)
        
        model = CustomFFNN()
        train_model(model, train_loader, val_loader, epochs=epochs)
        
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                outputs = model(xb)
                _, predicted = torch.max(outputs.data, 1)
                total += yb.size(0)
                correct += (predicted == yb).sum().item()
        
        val_accuracy = correct / total
        results[name] = val_accuracy
        print(f"\n{name} final validation accuracy: {val_accuracy:.4f}")
    
    return results

architectures = {
    "Original": [
        nn.Linear(1024, 512),
        nn.ReLU(),
        nn.Dropout(0.3),
        nn.Linear(512, 128),
        nn.ReLU(),
        nn.Linear(128, 10)
    ],
    "Deeper": [
        nn.Linear(1024, 768),
        nn.ReLU(),
        nn.Dropout(0.3),
        nn.Linear(768, 512),
        nn.ReLU(),
        nn.Dropout(0.3),
        nn.Linear(512, 256),
        nn.ReLU(),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 10)
    ],
    "Wider": [
        nn.Linear(1024, 1024),
        nn.ReLU(),
        nn.Dropout(0.4),
        nn.Linear(1024, 512),
        nn.ReLU(),
        nn.Linear(512, 10)
    ],
    "Narrower": [
        nn.Linear(1024, 256),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 10)
    ]
}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
arch_results = test_architectures(train_loader, val_loader, architectures)
````

Zkusme nějak optimalizovat, porovnáme  metody adam, adamw, sgd, rmsprop.

````python
def train_model_optimized(model, train_loader, val_loader, optimizer_type='adam', lr=0.001, epochs=15):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    
    if optimizer_type.lower() == 'adam':
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_type.lower() == 'adamw':
        optimizer = optim.AdamW(model.parameters(), lr=lr)
    elif optimizer_type.lower() == 'sgd':
        optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    elif optimizer_type.lower() == 'rmsprop':
        optimizer = optim.RMSprop(model.parameters(), lr=lr)
    else:
        raise ValueError("Unsupported optimizer type")
    
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=2, factor=0.5)
    
    best_val_accuracy = 0
    for epoch in range(epochs):
        model.train()
        running_loss = 0
        total_samples = 0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * xb.size(0)
            total_samples += xb.size(0)

        avg_loss = running_loss / total_samples
        print(f"Epoch {epoch+1}, Train Loss: {avg_loss:.4f}")

        model.eval()
        correct, total = 0, 0
        val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                outputs = model(xb)
                loss = criterion(outputs, yb)
                val_loss += loss.item() * xb.size(0)

                _, predicted = torch.max(outputs.data, 1)
                total += yb.size(0)
                correct += (predicted == yb).sum().item()
        
        val_accuracy = correct / total
        val_loss = val_loss / total
        print(f"Validation Loss: {val_loss:.4f}, Accuracy: {val_accuracy:.4f}")
        
        scheduler.step(val_accuracy)
        
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(model.state_dict(), 'best_model.pth')
    
    print(f"\nBest validation accuracy: {best_val_accuracy:.4f}")
    return best_val_accuracy

optimizers = ['adam', 'adamw', 'sgd', 'rmsprop']
optim_results = {}

for opt in optimizers:
    print(f"\nTesting optimizer: {opt}")
    print("-" * 50)
    model = FFNN()
    best_acc = train_model_optimized(model, train_loader, val_loader, optimizer_type=opt, lr=0.001)
    optim_results[opt] = best_acc
````

Vyzkoušíme různé regularizační techniky, včetně jejich kombinací.

````python
class RegularizedFFNN(nn.Module):
    def __init__(self, dropout_rate=0.3, use_batchnorm=False, l2_penalty=0.0):
        super(RegularizedFFNN, self).__init__()
        layers = []
        
        layers.append(nn.Linear(1024, 512))
        if use_batchnorm:
            layers.append(nn.BatchNorm1d(512))
        layers.append(nn.ReLU())
        layers.append(nn.Dropout(dropout_rate))
        
        layers.append(nn.Linear(512, 256))
        if use_batchnorm:
            layers.append(nn.BatchNorm1d(256))
        layers.append(nn.ReLU())
        layers.append(nn.Dropout(dropout_rate))
        
        layers.append(nn.Linear(256, 10))
        
        self.net = nn.Sequential(*layers)
        self.l2_penalty = l2_penalty
    
    def forward(self, x):
        return self.net(x)
    
    def l2_reg(self):
        l2_loss = 0.0
        for param in self.parameters():
            l2_loss += torch.norm(param, 2)
        return self.l2_penalty * l2_loss

def train_regularized_model(model, train_loader, val_loader, epochs=15):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.001)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=2, factor=0.5)
    
    best_val_accuracy = 0
    for epoch in range(epochs):
        model.train()
        running_loss = 0
        total_samples = 0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            outputs = model(xb)
            
            loss = criterion(outputs, yb)
            if hasattr(model, 'l2_reg'):
                loss += model.l2_reg()
            
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * xb.size(0)
            total_samples += xb.size(0)

        avg_loss = running_loss / total_samples
        print(f"Epoch {epoch+1}, Train Loss: {avg_loss:.4f}")

        model.eval()
        correct, total = 0, 0
        val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                outputs = model(xb)
                loss = criterion(outputs, yb)
                val_loss += loss.item() * xb.size(0)

                _, predicted = torch.max(outputs.data, 1)
                total += yb.size(0)
                correct += (predicted == yb).sum().item()
        
        val_accuracy = correct / total
        val_loss = val_loss / total
        print(f"Validation Loss: {val_loss:.4f}, Accuracy: {val_accuracy:.4f}")
        
        scheduler.step(val_accuracy)
        
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(model.state_dict(), 'best_regularized_model.pth')
    
    print(f"\nBest validation accuracy: {best_val_accuracy:.4f}")
    return best_val_accuracy

reg_configs = [
    {"name": "Base", "dropout_rate": 0.3, "use_batchnorm": False, "l2_penalty": 0.0},
    {"name": "More Dropout", "dropout_rate": 0.5, "use_batchnorm": False, "l2_penalty": 0.0},
    {"name": "BatchNorm", "dropout_rate": 0.3, "use_batchnorm": True, "l2_penalty": 0.0},
    {"name": "L2 Regularization", "dropout_rate": 0.3, "use_batchnorm": False, "l2_penalty": 0.001},
    {"name": "Combined", "dropout_rate": 0.4, "use_batchnorm": True, "l2_penalty": 0.0005}
]

reg_results = {}
for config in reg_configs:
    print(f"\nTesting regularization: {config['name']}")
    print("-" * 50)
    model = RegularizedFFNN(
        dropout_rate=config['dropout_rate'],
        use_batchnorm=config['use_batchnorm'],
        l2_penalty=config['l2_penalty']
    )
    best_acc = train_regularized_model(model, train_loader, val_loader)
    reg_results[config['name']] = best_acc
````

Nejlepších výsledků jsme dosáhli s použitím batch normalizace, ostatní metody nepřinesly výrazné zlepšení.

Můžeme tedy vytvořit nejlepší FFNN model s použitím batch normalizace a optimalizátoru Adam.

````python
class BestFFNN(nn.Module):
    def __init__(self):
        super(BestFFNN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            
            nn.Linear(128, 10)
        )
    
    def forward(self, x):
        return self.net(x)

def train_best_model(model, train_loader, val_loader, epochs=20):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    
    optimizer = optim.AdamW(model.parameters(), lr=0.001)
    
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=2
    )
    
    criterion = nn.CrossEntropyLoss()
    best_val_accuracy = 0
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0
        total_samples = 0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * xb.size(0)
            total_samples += xb.size(0)

        avg_loss = running_loss / total_samples
        print(f"Epoch {epoch+1}, Train Loss: {avg_loss:.4f}")

        # Validation
        model.eval()
        correct, total = 0, 0
        val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                outputs = model(xb)
                loss = criterion(outputs, yb)
                val_loss += loss.item() * xb.size(0)

                _, predicted = torch.max(outputs.data, 1)
                total += yb.size(0)
                correct += (predicted == yb).sum().item()
        
        val_accuracy = correct / total
        val_loss = val_loss / total
        print(f"Validation Loss: {val_loss:.4f}, Accuracy: {val_accuracy:.4f}")
        
        scheduler.step(val_accuracy)
      
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(model.state_dict(), 'best_model.pth')
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= 4:
                print("Early stopping triggered")
                break
    
    print(f"\nBest validation accuracy: {best_val_accuracy:.4f}")
    return best_val_accuracy

best_model = BestFFNN()
final_accuracy = train_best_model(best_model, train_loader, val_loader, epochs=20)
````

### Konvoluční neuronová síť

Na rozdíl od FNN konvoluční neuronová síť pracuje s 2D strukturou obrázků, zatímco FNN převádí vstup do 1D vektoru. Díky tomu CNN lépe zachycuje prostorové vztahy mezi pixely. Takže CNN používají filtry, které detekují lokální vzory – hrany, textury, tvary – což je ideální pro rozpoznávání oblečení.



Nyní vytvoříme CNN, kde na začátku převedeme data do tvaru (N, 1, 32, 32).Používáme max-pooling a dropout pro regularizaci.

````python
class AdaptedCNN(nn.Module):
    def __init__(self):
        super(AdaptedCNN, self).__init__()
        self.reshape = lambda x: x.view(-1, 1, 32, 32)
        
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        
        self.fc1 = nn.Linear(64 * 8 * 8, 128)
        self.fc2 = nn.Linear(128, 10)
        self.dropout = nn.Dropout(0.25)
        
    def forward(self, x):
        x = self.reshape(x)
        
        x = self.pool(nn.functional.relu(self.conv1(x)))
        x = self.pool(nn.functional.relu(self.conv2(x)))
        
        x = x.view(-1, 64 * 8 * 8)
        x = nn.functional.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x
````

Trénovací funkce, ve které použijeme optimalizátor Adam. Na konci vykreslíme grafy průběhu trénovací a validační ztráty i přesnosti.

````python
def train_model_with_plots(model, train_loader, val_loader, optimizer, criterion, epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []
    
    best_acc = 0
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()
        
        train_loss = train_loss / len(train_loader.dataset)
        train_acc = train_correct / train_total
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        
        model.eval()
        val_loss, val_correct, val_total = 0, 0, 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * images.size(0)
                
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
        
        val_loss = val_loss / len(val_loader.dataset)
        val_acc = val_correct / val_total
        val_losses.append(val_loss)
        val_accs.append(val_acc)
        
        print(f"Epoch {epoch+1}/{epochs}:")
        print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
        print(f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
        
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
    

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(val_accs, label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    print(f"Nejlepší validační přesnost: {best_acc:.4f}")
    return best_acc, train_losses, val_losses, train_accs, val_accs

model = AdaptedCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

best_acc, train_losses, val_losses, train_accs, val_accs = train_model_with_plots(
    model, train_loader, val_loader, optimizer, criterion, epochs=20
)
````
