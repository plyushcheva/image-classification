### SVM

Výhodou SVM je, že riziko přeučení je malé, protože model se snaží maximalizovat margin – vzdálenost mezi rozhodovací hranicí a nejbližšími vzorky.

Dobře funguje u vysokodimenzionálních dat, což se nám hodí.

Vyzkoušíme dvě jádrové funkce – lineární a RBF.

````python
linear_svm = SVC(kernel='linear', C=1, random_state=rd_seed)
linear_svm.fit(Xtrain, ytrain)
yval_pred_linear = linear_svm.predict(Xval)
print("Linear SVM Validation Accuracy:", accuracy_score(yval, yval_pred_linear))
````

````python
rbf_svm = SVC(kernel='rbf', C=1, random_state=rd_seed)
rbf_svm.fit(Xtrain, ytrain)
yval_pred_rbf = rbf_svm.predict(Xval)
print("RBF SVM Validation Accuracy:", accuracy_score(yval, yval_pred_rbf))
````

Pak přejdeme na *normalizaci* a napíšeme normalizační funkce.



````python
def apply_minmax_normalization(Xtrain, Xval, Xtest):
    scaler = MinMaxScaler()
    Xtrain_norm = scaler.fit_transform(Xtrain)
    Xval_norm = scaler.transform(Xval)
    Xtest_norm = scaler.transform(Xtest)
    return Xtrain_norm, Xval_norm, Xtest_norm

def apply_robust_normalization(Xtrain, Xval, Xtest):
    scaler = RobustScaler()
    Xtrain_norm = scaler.fit_transform(Xtrain)
    Xval_norm = scaler.transform(Xval)
    Xtest_norm = scaler.transform(Xtest)
    return Xtrain_norm, Xval_norm, Xtest_norm

def apply_standardization(Xtrain, Xval, Xtest):
    scaler = StandardScaler()
    Xtrain_norm = scaler.fit_transform(Xtrain)
    Xval_norm = scaler.transform(Xval)
    Xtest_norm = scaler.transform(Xtest)
    return Xtrain_norm, Xval_norm, Xtest_norm
````

````python
def run_svm_models(Xtrain_norm, Xval_norm, ytrain, yval, norm_name):
    print(f"\n Results using {norm_name} normalization:")

    # Linear
    linear_svm = SVC(kernel='linear', C=1, random_state=rd_seed)
    linear_svm.fit(Xtrain_norm, ytrain)
    yval_pred_linear = linear_svm.predict(Xval_norm)
    acc_linear = accuracy_score(yval, yval_pred_linear)
    print(f" Linear SVM Validation Accuracy: {acc_linear:.4f}")
    
    # RBF
    rbf_svm = SVC(kernel='rbf', C=1, random_state=rd_seed)
    rbf_svm.fit(Xtrain_norm, ytrain)
    yval_pred_rbf = rbf_svm.predict(Xval_norm)
    acc_rbf = accuracy_score(yval, yval_pred_rbf)
    print(f" RBF SVM Validation Accuracy:    {acc_rbf:.4f}")
````

Otestujeme každou možnost.

````python
Xtrain_minmax, Xval_minmax, Xtest_minmax = apply_minmax_normalization(Xtrain, Xval, Xtest)
run_svm_models(Xtrain_minmax, Xval_minmax, ytrain, yval, "Min-Max")

Xtrain_robust, Xval_robust, Xtest_robust = apply_robust_normalization(Xtrain, Xval, Xtest)
run_svm_models(Xtrain_robust, Xval_robust, ytrain, yval, "Robust")

Xtrain_std, Xval_std, Xtest_std = apply_standardization(Xtrain, Xval, Xtest)
run_svm_models(Xtrain_std, Xval_std, ytrain, yval, "Standard")
````

Nejlepší výsledek tedy dostáváme u RBF SVM, a to jak s použitím MinMax normalizace, tak bez ní.

Přejdeme na **Ladění hyperparametrů:**

````python
param_grid = {
    'C': range(1, 20),
    'gamma': ['scale', 'auto'],
    'kernel': ['rbf', 'linear']
}

grid = ParameterGrid(param_grid)

val_metric = []
for params in grid:
    clfSVM = SVC(**params, random_state=rd_seed)
    clfSVM.fit(Xtrain_minmax, ytrain)
    val_metric.append(metrics.accuracy_score(yval, clfSVM.predict(Xval_minmax)))
````

````python
best_params_svm = grid[np.argmax(val_metric)]
best_accuracy_svm = max(val_metric)

plt.figure(figsize=(20,6))
plt.plot(val_metric,  'ob-')
plt.xlabel('Hyperparameter index')
plt.ylabel('Accuracy')
plt.legend(['Validation'])
plt.title('Validation accuracies for different hyperparameters')
plt.show()
print(f"Best params {best_params_svm} with validation accuracy {best_accuracy_svm:.4f}.")
````

Jen mírně jsme zlepšili výsledek a dosáhli accuracy 0.9733.

### Naivní Bayesův klasifikátor

Zkusíme využít Naivní Bayesův klasifikátor – velkou výhodou tohoto modelu je výpočetní nenáročnost.
Porovnáváme výsledky GaussianNB, BernoulliNB a MultinomialNB:

````python
Xtrain_nb, Xval_nb, Xtest_nb = Xtrain, Xval, Xtest

gnb = GaussianNB()
gnb.fit(Xtrain_nb, ytrain)
yval_gnb = gnb.predict(Xval_nb)
acc_gnb = accuracy_score(yval, yval_gnb)

bnb = BernoulliNB()
bnb.fit(Xtrain_nb, ytrain)
yval_bnb = bnb.predict(Xval_nb)
acc_bnb = accuracy_score(yval, yval_bnb)

mnb = MultinomialNB()
mnb.fit(Xtrain, ytrain)
yval_mnb = mnb.predict(Xval)
acc_mnb = accuracy_score(yval, yval_mnb)

print(f"GaussianNB Validation Accuracy:   {acc_gnb:.4f}")
print(f"BernoulliNB Validation Accuracy:  {acc_bnb:.4f}")
print(f"MultinomialNB Validation Accuracy:{acc_mnb:.4f}")
````

Nejlepší výsledek máme u MultinomialNB.

Bayesův klasifikátor s multinomickým rozdělením se obecně používá se vstupními daty, která reprezentují nějakou frekvenci nebo intenzitu (např. četnost slov), což ale příliš neodpovídá našemu datasetu s obrázky.

GaussianNB je vhodný pro spojitá data, což pixelové hodnoty jsou.
Nicméně: pixelové hodnoty nejsou vždy normálně rozložené, což porušuje předpoklad normality pro GaussianNB.

BernoulliNB se používá pro binární data, např. černý/bílý pixel.
Takže v našem případě se příliš nehodí, protože ignoruje jemné rozdíly v odstínech.



**Normalizace dat** 

Normalizace se hodí spíš pro Bayesův klasifikátor s Gaussovským rozdělením, proto ji aplikujeme primárně pro GaussianNB.
Pro BernoulliNB a MultinomialNB ji zkoušíme spíše ze zvědavosti.

````python
def apply_minmax_normalization(Xtrain, Xval, Xtest):
    scaler = MinMaxScaler()
    return scaler.fit_transform(Xtrain), scaler.transform(Xval), scaler.transform(Xtest)

def apply_standardization(Xtrain, Xval, Xtest):
    scaler = StandardScaler()
    return scaler.fit_transform(Xtrain), scaler.transform(Xval), scaler.transform(Xtest)

def apply_robust_normalization(Xtrain, Xval, Xtest):
    scaler = RobustScaler()
    return scaler.fit_transform(Xtrain), scaler.transform(Xval), scaler.transform(Xtest)

def apply_binarization(Xtrain, Xval, Xtest, threshold=0):
    binarizer = Binarizer(threshold=threshold)
    return binarizer.fit_transform(Xtrain), binarizer.transform(Xval), binarizer.transform(Xtest)
````

Použijeme několik normalizačních funkcí.

*minmax*:

````python
Xtrain_gnb_minmax, Xval_gnb_minmax, Xtest_gnb_minmax = apply_minmax_normalization(Xtrain, Xval, Xtest)
gnb_minmax = GaussianNB()
gnb_minmax.fit(Xtrain_gnb_minmax, ytrain)
acc_gnb_minmax = accuracy_score(yval, gnb_minmax.predict(Xval_gnb_minmax))
print(f"Accuracy GaussianNB + MinMaxScaler: {acc_gnb_minmax:.4f}")

Xtrain_mnb_minmax, Xval_mnb_minmax, Xtest_mnb_minmax = apply_minmax_normalization(Xtrain, Xval, Xtest)
mnb_minmax = MultinomialNB()
mnb_minmax.fit(Xtrain_mnb_minmax, ytrain)
acc_mnb_minmax = accuracy_score(yval, mnb_minmax.predict(Xval_mnb_minmax))
print(f"Accuracy MultinomialNB + MinMaxScaler: {acc_mnb_minmax:.4f}")
````

Přesnost u GaussianNB se nezměnila, u MultinomialNB se mírně zhoršila.
Zkusíme provést binarizaci pro GaussianNB a BernoulliNB:

````python
Xtrain_gnb_bin, Xval_gnb_bin, Xtest_gnb_bin = apply_binarization(Xtrain, Xval, Xtest)
gnb_bin = GaussianNB()
gnb_bin.fit(Xtrain_gnb_bin, ytrain)
acc_gnb_bin = accuracy_score(yval, gnb_bin.predict(Xval_gnb_bin))
print(f"Accuracy GaussianNB + Binarizer: {acc_gnb_bin:.4f}")

Xtrain_bnb_bin, Xval_bnb_bin, Xtest_bnb_bin = apply_binarization(Xtrain, Xval, Xtest)
bnb_bin = BernoulliNB()
bnb_bin.fit(Xtrain_bnb_bin, ytrain)
acc_bnb_bin = accuracy_score(yval, bnb_bin.predict(Xval_bnb_bin))
print(f"Accuracy BernoulliNB + Binarizer: {acc_bnb_bin:.4f}")
````

Získáváme lepší výsledek pro GaussianNB.

Dále zkusíme u GaussianNB ještě standardizaci a robust scaling.



````python
Xtrain_gnb_std, Xval_gnb_std, Xtest_gnb_std = apply_standardization(Xtrain, Xval, Xtest)
gnb_std = GaussianNB()
gnb_std.fit(Xtrain_gnb_std, ytrain)
acc_gnb_std = accuracy_score(yval, gnb_std.predict(Xval_gnb_std))
print(f"Accuracy GaussianNB + StandardScaler: {acc_gnb_std:.4f}")

Xtrain_gnb_robust, Xval_gnb_robust, Xtest_gnb_robust = apply_robust_normalization(Xtrain, Xval, Xtest)
gnb_robust = GaussianNB()
gnb_robust.fit(Xtrain_gnb_robust, ytrain)
acc_gnb_robust = accuracy_score(yval, gnb_robust.predict(Xval_gnb_robust))
print(f"Accuracy GaussianNB + RobustScaler: {acc_gnb_robust:.4f}")
````

Nejlepší výsledek máme u MultinomialNB a GaussianNB + Binarizace.

Přejdeme na **Ladění hyperparametrů:**

*GaussianNB:*

````python
var_smoothing_values = np.arange(1e-9, 1e-5, 1e-8)
accuracies_gnb = []

for vs in var_smoothing_values:
    gnb = GaussianNB(var_smoothing=vs)
    gnb.fit(Xtrain, ytrain)
    y_pred = gnb.predict(Xval)
    acc = accuracy_score(yval, y_pred)
    accuracies_gnb.append(acc)

best_vs = var_smoothing_values[np.argmax(accuracies_gnb)]
best_acc_gnb = max(accuracies_gnb)

plt.figure(figsize=(10, 5))
plt.plot(var_smoothing_values, accuracies_gnb, marker='o')
plt.xscale('log')
plt.title('Validation Accuracy - GaussianNB (var_smoothing)')
plt.xlabel('var_smoothing')
plt.ylabel('Accuracy')
plt.show()

print(f"Best GaussianNB var_smoothing: {best_vs:.2e} with accuracy: {best_acc_gnb:.4f}")
````

Accuracy u GaussianNB se jen lehce zlepšila.

*BernoulliNB:*

````python
param_grid = {
    'alpha': np.arange(1e-9, 3, 0.5),
    'binarize': np.arange(1e-9, 5, 0.5)
}
grid = list(ParameterGrid(param_grid))

results = []
for params in grid:
    bnb = BernoulliNB(**params)
    bnb.fit(Xtrain, ytrain)
    y_pred = bnb.predict(Xval)
    acc = accuracy_score(yval, y_pred)
    results.append({
        'alpha': params['alpha'],
        'binarize': params['binarize'],
        'accuracy': acc
    })
best_result = max(results, key=lambda x: x['accuracy'])
val_metric = [r['accuracy'] for r in results]

plt.figure(figsize=(20,8))
plt.plot(val_metric, 'ob-')
plt.xlabel('Hyperparameter index')
plt.ylabel('Accuracy')
plt.legend(['Validation'])
plt.title('Validation accuracies for different hyperparameters')
plt.show()

print(f"Best BernoulliNB params: alpha={best_result['alpha']}, binarize={best_result['binarize']} with accuracy={best_result['accuracy']:.4f}")
````

Výsledek u BernoulliNB se zlepšil.

*MultinomialNB:*

````python
param_grid = {
    'alpha': np.arange(1e-9, 3, 0.5)
}
grid = list(ParameterGrid(param_grid))

results = []
for params in grid:
    mnb = MultinomialNB(**params)
    mnb.fit(Xtrain, ytrain)
    y_pred = mnb.predict(Xval)
    acc = accuracy_score(yval, y_pred)
    results.append({
        'alpha': params['alpha'],
        'accuracy': acc
    })

best_result = max(results, key=lambda x: x['accuracy'])

val_metric = [r['accuracy'] for r in results]
plt.figure(figsize=(10,4))
plt.plot(val_metric, 'ob-')
plt.xlabel('Hyperparameter index')
plt.ylabel('Accuracy')
plt.legend(['Validation'])
plt.title('Validation accuracies for different alpha values (MultinomialNB)')
plt.show()

print(f"Best MultinomialNB params: alpha={best_result['alpha']} with accuracy={best_result['accuracy']:.4f}")
````

Ladění hyperparametrů nezměnilo accuracy u MultinomialNB.

Z toho vyplývá, že ze všech tří variant má nejlepší accuracy BernoulliNB.

### LDA

Přejde na Lineární diskriminační analýzu:

````python
lda = LinearDiscriminantAnalysis()
lda.fit(Xtrain, ytrain)
acc_lda = accuracy_score(yval, lda.predict(Xval))
print(f"Accuracy LDA: {acc_lda:.4f}")
````

Škálování nemá vliv na Lineární diskriminační analýzu, ale zkusíme MinMax a standardizaci pro ukázku:

````python
scaler_std = StandardScaler()
Xtrain_std = scaler_std.fit_transform(Xtrain)
Xval_std = scaler_std.transform(Xval)

lda_std = LinearDiscriminantAnalysis()
lda_std.fit(Xtrain_std, ytrain)
acc_lda_std = accuracy_score(yval, lda_std.predict(Xval_std))
print(f"Accuracy LDA + StandardScaler: {acc_lda_std:.4f}")

scaler_mm = MinMaxScaler()
Xtrain_mm = scaler_mm.fit_transform(Xtrain)
Xval_mm = scaler_mm.transform(Xval)

lda_mm = LinearDiscriminantAnalysis()
lda_mm.fit(Xtrain_mm, ytrain)
acc_lda_mm = accuracy_score(yval, lda_mm.predict(Xval_mm))
print(f"Accuracy LDA + MinMaxScaler: {acc_lda_mm:.4f}")
````

Vidíme, že hodnota se nezměnila.

Přejdeme na ladění **hyperparametrů**, abychom vylepšili výsledek.

````python
param_grid = {
    'solver': [ 'lsqr', 'eigen'],
    'shrinkage': [None, 'auto', 0.1, 0.2, 0,3, 0.5, 0.6, 1],
    'n_components': [1, 2, 3]
}

grid = list(ParameterGrid(param_grid))
results = []

for params in grid:
    try:
        lda = LinearDiscriminantAnalysis(
            solver=params['solver'],
            shrinkage=params['shrinkage'] if params['solver'] != 'svd' else None,
            n_components=params['n_components']
        )
        lda.fit(Xtrain, ytrain)
        y_pred = lda.predict(Xval)
        acc = accuracy_score(yval, y_pred)

        results.append({
            'solver': params['solver'],
            'shrinkage': params['shrinkage'],
            'n_components': params['n_components'],
            'accuracy': acc
        })

    except Exception as e:
        continue


results_sorted = sorted(results, key=lambda x: x['accuracy'], reverse=True)
best = results_sorted[0]
````

````python
plt.figure(figsize=(14, 6))
plt.plot([f"{r['solver']}, sh={r['shrinkage']}, n={r['n_components']}" for r in results],
         [r['accuracy'] for r in results], marker='o', linestyle='--', color='teal')
plt.xticks(rotation=90)
plt.title("LDA Hyperparameter Tuning - Validation Accuracy")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()
````

````python
print(f"Best LDA Params: solver={best['solver']}, shrinkage={best['shrinkage']}, n_components={best['n_components']}")
print(f"Best Validation Accuracy: {best['accuracy']:.4f}")
````

Našli jsme vhodné parametry – nyní LDA accuracy = 0.9667.

### Redukce dimenzionality

#### 1. PCA

Pojďme zkusit vylepšit modely pomoci PCA.

* ##### SVM

````python
search_space_svm = {
    'pca_dim': range(10, 15),
    'C': range(1, 10),
    'gamma': ['scale', 'auto'],
    'kernel': ['rbf', 'linear']
}

all_settings_svm = list(ParameterGrid(search_space_svm))
acc_scores_svm = []

for config in all_settings_svm:
    pca_model = PCA(n_components=config['pca_dim'], random_state=rd_seed)
    reduced_Xtrain = pca_model.fit_transform(Xtrain_minmax)
    reduced_Xval = pca_model.transform(Xval_minmax)

    model_svm = SVC(C=config['C'], gamma=config['gamma'], kernel=config['kernel'], random_state=rd_seed)
    model_svm.fit(reduced_Xtrain, ytrain)

    preds = model_svm.predict(reduced_Xval)
    acc_scores_svm.append(accuracy_score(yval, preds))

optimal_idx_svm = np.argmax(acc_scores_svm)
top_svm_config = all_settings_svm[optimal_idx_svm]
top_svm_acc = acc_scores_svm[optimal_idx_svm]

print(f"SVM + PCA config: {top_svm_config}")
print(f"Best validation accuracy: {top_svm_acc:.4f}")
````

PCA se nezlepšilo výsledek SVM.

* ##### Naivní Bayesův klasifikátor

Zkusíme použít PCA pro GaussianNB a BernoulliNB, protože MultinomialNB se pro PCA příliš nehodí – očekává počty nebo frekvence výskytu rysů, tedy nezáporná čísla.

PCA ale produkuje záporné i spojité hodnoty, což porušuje předpoklady MultinomialNB.

*GaussianNB:*

````python
search_space_nb = {
    'pca_dim': range(1, 20),
    'var_smoothing': np.arange(1e-9, 1e-7, 1e-8),
    'scaler': [Binarizer()]
}

configs_nb = list(ParameterGrid(search_space_nb))
scores_nb = []

for cfg in configs_nb:
    norm = cfg['scaler'].fit(Xtrain)
    Xt = norm.transform(Xtrain)
    Xv = norm.transform(Xval)

    pca_nb = PCA(n_components=cfg['pca_dim'], random_state=rd_seed).fit(Xt)
    Xt_pca = pca_nb.transform(Xt)
    Xv_pca = pca_nb.transform(Xv)

    gnb = GaussianNB(var_smoothing=cfg['var_smoothing'])
    gnb.fit(Xt_pca, ytrain)
    preds_nb = gnb.predict(Xv_pca)
    scores_nb.append(accuracy_score(yval, preds_nb))

best_nb = configs_nb[np.argmax(scores_nb)]
acc_best_nb = max(scores_nb)

plt.figure(figsize=(18, 5))
plt.scatter(range(len(scores_nb)), scores_nb, c='green', label='Accuracies')
plt.axhline(y=acc_best_nb, color='black', linestyle='--', label='Max Accuracy')
plt.title('GaussianNB + PCA (Validation Accuracy)')
plt.xlabel('Parameter Combination Index')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

print(f"GaussianNB + PCA config: {best_nb}")
print(f"Best validation accuracy: {acc_best_nb:.4f}")
````

PCA zlepšilo výsledek pro GaussianNB

*BernoulliNB:*

````python
search_space_bnb = {
    'pca_dim': range(15, 20),
    'alpha': np.arange(0.01, 1, 0.5),
    'binarize': np.arange(0.01, 1.5, 0.1),
    'scaler': [Binarizer()]
}

configs_bnb = list(ParameterGrid(search_space_bnb))
scores_bnb = []

for cfg in configs_bnb:
    scaler = cfg['scaler'].fit(Xtrain)
    Xt = scaler.transform(Xtrain)
    Xv = scaler.transform(Xval)

    pca_bnb = PCA(n_components=cfg['pca_dim'], random_state=rd_seed).fit(Xt)
    Xt_pca = pca_bnb.transform(Xt)
    Xv_pca = pca_bnb.transform(Xv)

    clf_bnb = BernoulliNB(alpha=cfg['alpha'], binarize=cfg['binarize'])
    clf_bnb.fit(Xt_pca, ytrain)
    preds_bnb = clf_bnb.predict(Xv_pca)
    scores_bnb.append(accuracy_score(yval, preds_bnb))

best_bnb = configs_bnb[np.argmax(scores_bnb)]
acc_best_bnb = max(scores_bnb)

plt.figure(figsize=(18, 5))
plt.scatter(range(len(scores_bnb)), scores_bnb, c='orange', label='Accuracies')
plt.axhline(y=acc_best_bnb, color='black', linestyle='--', label='Max Accuracy')
plt.title('BernoulliNB + PCA (Validation Accuracy)')
plt.xlabel('Parameter Combination Index')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

print(f"BernoulliNB + PCA config: {best_bnb}")
print(f"Best validation accuracy: {acc_best_bnb:.4f}")
````

Výsledek pro BernoulliNB se nezlepšil

* ##### LDA

````python
search_space_lda = {
    'pca_dim': range(10, 15),
    'solver': ['lsqr', 'eigen'],
    'shrinkage': np.round(np.arange(0.01, 1.0, 0.01), 2),
    'lda_components': range(1, min(len(np.unique(ytrain)), Xtrain.shape[1])),
    'scaler': [MinMaxScaler()]
}

grid_lda = list(ParameterGrid(search_space_lda))
accuracy_results_lda = []

for cfg in grid_lda:
    scaled = cfg['scaler'].fit_transform(Xtrain_minmax)
    scaled_val = cfg['scaler'].transform(Xval_minmax)

    pca_proj = PCA(n_components=cfg['pca_dim'], random_state=rd_seed).fit(scaled)
    Xtr = pca_proj.transform(scaled)
    Xv = pca_proj.transform(scaled_val)

    try:
        lda_model = LinearDiscriminantAnalysis(solver=cfg['solver'],
                                               shrinkage=cfg['shrinkage'],
                                               n_components=cfg['lda_components'])
        lda_model.fit(Xtr, ytrain)
        preds = lda_model.predict(Xv)
        score = accuracy_score(yval, preds)
    except:
        score = 0
    accuracy_results_lda.append(score)

best_index_lda = np.argmax(accuracy_results_lda)
best_setup_lda = grid_lda[best_index_lda]
best_score_lda = accuracy_results_lda[best_index_lda]

print(f"LDA + PCA config: {best_setup_lda}")
print(f"Best validation accuracy: {best_score_lda:.4f}")
````

PCA mírně zhoršilo výsledek u LDA

#### Použijeme natrénovaný LDA model pro zobrazení výsledků v podobě obrázků.

Pro každou třídu vezmeme její střední hodnotu a kovarianční matici z modelu.
Vygenerujeme 5 vzorků pomocí np.random.multivariate_normal.

````python
lda_gen = LinearDiscriminantAnalysis(solver='lsqr', shrinkage=0.6, n_components=1)
lda_gen.fit(Xtrain, ytrain)

classes = lda_gen.classes_
means = lda_gen.means_
cov = lda_gen.covariance_

if cov.ndim == 2:
    cov_matrix = cov
else:
    cov_matrix = np.diag(cov)

generated_data = []

for idx, label in enumerate(classes):
    samples = np.random.multivariate_normal(mean=means[idx], cov=cov_matrix, size=5)
    generated_data.append(samples)

fig, axes = plt.subplots(len(classes), 5, figsize=(10, 8))
fig.suptitle('LDA - Vygenerovaná data', fontsize=16)

for i, class_samples in enumerate(generated_data):
    for j in range(5):
        ax = axes[i, j]
        ax.imshow(class_samples[j].reshape(28, 28), cmap='gray')
        ax.axis('off')
        if j == 0:
            ax.set_ylabel(f"Třída {classes[i]}", fontsize=12)

plt.tight_layout()
plt.show()
````

Generovaná data mají celkovou strukturu podobnou originálu – je vidět obrys oblečení, je patrné, kde je svrchní oděv.

Detaily obrázků jsou však rozmazané – LDA předpokládá lineární rozdělení, a není schopna zachytit nelineární strukturu obrazu.

Vzhledem k tomu, že používáme sdílenou kovarianci, všechny třídy mají stejnou variabilitu, což není realistické.
Pro lepší vizuální kvalitu by bylo potřeba použít nelineární generativní modely.
